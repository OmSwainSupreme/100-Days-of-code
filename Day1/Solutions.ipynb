```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 200

df = pd.DataFrame({
    "EmployeeID": np.arange(1001, 1001+n),
    "Name": np.random.choice(["Amit", "Sara", "John", "Priya", "Ali", "Ravi"], n),
    "Age": np.random.randint(21, 60, n),
    "Department": np.random.choice(["IT", "HR", "Finance", "Sales"], n),
    "Salary": np.random.randint(20000, 120000, n),
    "Experience": np.random.randint(0, 20, n),
    "City": np.random.choice(["Delhi", "Mumbai", "Bangalore", "Kolkata"], n),
    "JoiningYear": np.random.randint(2005, 2023, n),
    "PerformanceScore": np.random.randint(1, 10, n),
    "Bonus": np.random.randint(5000, 30000, n),
    "ProjectsHandled": np.random.randint(1, 15, n),
    "Attrition": np.random.choice(["Yes", "No"], n)
})
# Randomly introduce nulls
for col in ["Age", "Salary", "Bonus"]:
    df.loc[df.sample(frac=0.1).index, col] = np.nan
# Add salary outliers
outlier_indices = df.sample(5).index
df.loc[outlier_indices, "Salary"] = df["Salary"] * 5

# Add bonus outliers
outlier_indices = df.sample(5).index
df.loc[outlier_indices, "Bonus"] = df["Bonus"] * 4
df2 = pd.DataFrame({
    "EmployeeID": np.random.choice(df["EmployeeID"], 150),
    "BankAccount": np.random.randint(10000000, 99999999, 150),
    "LoanAmount": np.random.randint(10000, 500000, 150),
    "CreditScore": np.random.randint(300, 900, 150)
})
df2 = pd.DataFrame({
    "EmployeeID": np.random.choice(df["EmployeeID"], 150),
    "BankAccount": np.random.randint(10000000, 99999999, 150),
    "LoanAmount": np.random.randint(10000, 500000, 150),
    "CreditScore": np.random.randint(300, 900, 150)
})
df.to_csv("employees.csv", index=False)
df2.to_csv("financials.csv", index=False)
```

```python
import pandas as pd
df1 = pd.read_csv("employees.csv")
df2 = pd.read_csv("financials.csv")
```

```python
pd.merge(df1,df2,on="EmployeeID")
pd.merge(df1,df2,how="inner")
pd.merge(df1,df2,how="right")
pd.merge(df1,df2,how="left")
pd.merge(df1,df2,how="outer")

pd.concat([df1,df2],axis=1)
pd.concat([df1,df2],axis=0)

df1.join(df2,lsuffix="LEFT",rsuffix="RIGHT")

pd.concat([df1,df2],ignore_index=True)

pd.merge(df1,df2,how="inner")
```
